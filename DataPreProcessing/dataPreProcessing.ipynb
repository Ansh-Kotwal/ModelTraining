```python
dataframe['Date'] = dataframe['time'].dt.date
wdata = dataframe.groupby('Date').mean()
wdata = wdata.drop(columns=['time'])
wdata['Date'] = dataframe['Date'].unique()
wdata['Date'] = pd.to_datetime(wdata["Date"])
wdata
```
